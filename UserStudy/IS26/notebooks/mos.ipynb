```python
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
csv_df = './results/mos.csv'
csv_df = pd.read_csv(csv_df)

# drop columns that doesn't contain Answer
csv_df = csv_df.drop(columns=[col for col in csv_df.columns if 'Answer' not in col])
```

```python
def parse_result(row):
    result = {}
    for col in csv_df.columns:
        data_name = col.split('Answer.MOS_')[1]
        model = data_name.split('_')[0]
        utternace = data_name.replace(model + '_', '')
        result[(model, utternace)] = row[col]
    return result
```

```python
result = defaultdict(list)
for iter, row in csv_df.iterrows():
    _result = parse_result(row)
    for key, value in _result.items():
        result[key[0]].append(value)
```

```python
for model in ['slt24', 'DarkStream', 'GenVC-small', 'TVTSyn', 'Ours']:
    if model in ['PT', 'PR']:
        continue
    print(f'{model}: {np.mean(result[model]):.2f}')
```
