```python
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
csv_df = './results/intelligibility.csv'
csv_df = pd.read_csv(csv_df)

# drop columns that doesn't contain Answer
csv_df = csv_df.drop(columns=[col for col in csv_df.columns if 'Answer_overall' not in col])
print(csv_df.shape)
```

```python
def parse_result(row):
    result = {}
    mapping = {
        'TAO': 0,
        'OAT': 1,
    }
    
    for col in csv_df.columns:
        data_name = col.split('Answer.INT_')[1]
        type = data_name.split('_')[0] # TAO: A is baseline and B is our model; OAT: A is our model and B is baseline
        utternace = data_name.replace(type + '_', '').replace('_Answer_overall', '')
        if type == 'TAO':
            result[utternace] = mapping['TAO'] if row[col] == 1 else mapping['OAT']
        elif type == 'OAT':
            result[utternace] = mapping['OAT'] if row[col] == 1 else mapping['TAO']
    return result
```

```python
result = defaultdict(list)
for iter, row in csv_df.iterrows():
    _result = parse_result(row)

    for key, value in _result.items():
        result[key].append(value)
```

```python
total_number = []
for utt in result:
    total_number.append(result[utt])
total_number = np.array(total_number)
total_number = total_number.flatten()
```

```python
print(total_number.mean())
```
